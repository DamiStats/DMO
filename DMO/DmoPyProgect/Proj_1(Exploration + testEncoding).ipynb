# Descrizione Dataset

Primo Anno Coorti 2010-2016 per 2 CdS della Scuola di SMFN

Il data set contiene record relativi a studenti di 7 coorti e 2 CdS della Scuola di SMFN. 
Ogni record ha i seguenti attributi: degree, cohort, gender, school, test, total_cfu, total_cfu_g, avg_g,
math_course_g, math_course_d 

In particolare, per ogni coorte si hanno i risultati della produttivita' fino a dicembre dell'anno successivo all'anno di immatricolazione:

immatricolati coorte 2010 - produttivita' fino a dicembre 2011
immatricolati coorte 2011 - produttivita' fino a dicembre 2012
immatricolati coorte 2012 - produttivita' fino a dicembre 2013
immatricolati coorte 2013 - produttivita' fino a dicembre 2014
immatricolati coorte 2014 - produttivita' fino a dicembre 2015
immatricolati coorte 2015 - produttivita' fino a dicembre 2016
immatricolati coorte 2016 - produttivita' fino a dicembre 2017

Legenda per school, la scuola di provenienza

LS = Liceo Scientifico
LC = Liceo Classico
IT = Istituto Tecnico Industriale
TC = Istituto Tecnico Commerciale
IP = Istituto Professionale
AL, XX, etc = Altro


Nota sul test (test di autovalutazione):
- negli anni 2010-2015, il test di autovalutazione della Scuola di SMFN era costituito da un questionario con 25 domande: 
ogni risposta corretta era valutata 1 e ogni risposta sbagliata o non data era valutata 0; il test risultava superato con un punteggio >=12.
- dal 2016, il test di autovalutazione della Scuola di Scienze e' costituito da un questionario con 20 domande:
ogni risposta corretta viene pesata con valore 1, ogni risposta sbagliata con -0.25 e ogni risposta non data con 0; il test risulta superato con punteggio >=8.


Nota sui crediti:
"total_cfu_g": sono crediti che corrispondono ad esami con attribuzione di voto, eventualmente anche riconoscimenti da altri CdS.
"total_cfu": comprendono i crediti con voto, quelli di esami per i quali non e' previsto un voto e quelli di eventuali 
riconoscimenti da altri CdS senza riattribuzione del voto.

"avg_g" e' il voto medio degli esami sostenuti

"math_course_g" e' il voto preso all'esame con contenuti di analisi matematica del I anno (se 0 significa che l'esame non e' stato sostenuto)
"math_course_d" e' la data dell'esame di cui al punto precedente

## La variabile test in relazione alla variabile cohort
Nota sul test (test di autovalutazione):
- negli anni 2010-2015, il test di autovalutazione della Scuola di SMFN era costituito da un questionario con 25 domande: 
ogni risposta corretta era valutata 1 e ogni risposta sbagliata o non data era valutata 0; il test risultava superato con un punteggio >=12.
- dal 2016, il test di autovalutazione della Scuola di Scienze e' costituito da un questionario con 20 domande:
ogni risposta corretta viene pesata con valore 1, ogni risposta sbagliata con -0.25 e ogni risposta non data con 0; il test risulta superato con punteggio >=8.


```python
# Calcolo il numero di studenti appartenenti alla e cohort 2010-2015 e quelli appartenenti a 2016
somma10_15 = 0
somma16 = 0
for i in df.cohort:
    if i != 2016:
        somma10_15= i/i + somma10_15
    else:
        somma16 = i/i + somma16
        
```

```python
print("N. studenti cohorte 2010-2015 = " + str(int(somma10_15)))
print("N. studenti cohorte 2016 = " + str(int(somma16)))
(somma10_15 + somma16) == len(df.cohort)
```

```python
# Calcolo le posizioni degli studenti di cohorte 10_15 e degli studenti di cohorte 16
posizione10_15 = []
posizione16 = []
for counter,value in enumerate(df.cohort):
    if value != 2016:
        #print(counter,value)
        posizione10_15.append(counter)
    else:
        #print(counter,value)
        posizione16.append(counter)

```

```python
# valuto le statistiche dei due gruppi di valutazione per stabilire che NORMALIZZAZIONE EFFETTUARE

print(summary(df.test[posizione10_15]))
print(summary(df.test[posizione16]))
print("FAI UN INTERVALLO DI CONFIDENZA")
```

```python
# dato che al passo precendete ho visto che gli studenti si sono classificati in media con una valutazione
# appena sufficiente per la modalità di valutazione (10-15) e con una valutazione insufficiente per l'anno 2016
# ho ritenuto opportuno dicotomizzare l'esito con 0, 1 per indicare l'avvenuta sufficienza o meno del test
#costruisco una variabile indicatrice dell'esito (successo = 1 se passato al test), (successo = 0 se non passato al test)
# del test tenendo conto delle due modalità di valutazione

i = 0 #contatore
esito = []
testNorm = []
for j in df.cohort:
    if j== 2016 and df.test[i] >= 8:
        esito.append(1)
    elif j== 2016 and df.test[i] < 8:
        esito.append(0)
    elif j != 2016 and df.test[i] >=12:
        esito.append(1)
    else:
        esito.append(0)
    print(str(esito[i]) +" "+ str(df.cohort[i]) +"  " + str(df.test[i]) ) 
    
    i = i +1
esito
        
```

```python
# aGGIUNGO LA VARIABILE ESITO CREATA SOPRA AL DF
# add new column using square bracket notation
df['EsitoTest'] = esito
df
```
