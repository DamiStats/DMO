# ESERCIZIO 1
Scrivere una funzione che dato un punteggio, ritorna una stringa  il voto del punteggio  seguendo questo schema:



```python
def voto(score):
    voto = '?'
    if score >= 90:
        voto= 'A'
    elif score >= 80 and score <90:
        voto = 'B'
    elif score >= 70 and score < 80:
        voto = 'C'
    elif score >= 60 and score < 70:
        voto = 'D'
    else:
        voto='F'
    return(voto)
        
```

```python

for i in range(50, 101, 5):
    print(str(i) + str(voto(i)))
```

# Esercizio 3

Scrivere una funzione chiamata is_even(n) che prende un intero come argomento e ritorna True se largomento  un numero pari and False se  dispari odd.

```python
def is_even(x):
    if x%2 == 0:
        return True
    else:
        return False
```

```python
def is_even_in_one_row(x):
    return x%2 == 0
```

```python

for i in range(1,11):
    print(str(i) + "  " + str(is_even(i)) + "  " + str(i) + "  " + str(is_even_in_one_row(i)))
```

Scrivere la funzione is_prime(n) che ritorna True quando n  primo e False altrimenti.

```python
def is_prime(n):
    if n%2 == 0 or n%3 == 0:
        return False
    else:
        return True
```

```python
def is_prime_in_one_row(n):
    return (n%2) == 0 or (n%3) == 0
```
