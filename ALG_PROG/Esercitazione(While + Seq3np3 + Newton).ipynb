# Il METODO NEWTON


## Metodo Newton attraverso il ciclo for

```python
def NewtonFor(Num, nIter):
    approx = 0.5 * Num    # x = approx
    for i in range(nIter):
        better = (1/2) * (approx + Num/approx) #f(x) = better
        approx = better
    return(better)
```

```python
print(NewtonFor(5,1))
print(NewtonFor(5,2))
print(NewtonFor(5,5))  
print(NewtonFor(5,10))
print("NB: La terza chiamata  uguale alla quarta")
```

# Metodo Newton attraverso ciclo while

```python
def NewtonWhile(Num):
    approx = 0.5 * Num
    better = (1/2)*(approx + Num/approx)
    while approx != better:
        approx = better
        better = (1/2)*(approx + Num/approx)
    return(better)
```

```python
print(NewtonWhile(5))
print(5**0.5)
```

# Scrivere una funzione print_triangular_numbers(n) che stampa i primi n numeri triangolari (Cercare sul web cosa  un numero triangolare). Una chiamata a print_triangular_numbers(5) produrrebbe il seguente risultato:

```python
def print_triangula_numbers(n):
    for i in range(n+1):
        operazione = int((i*(i+1))/2)
        print(str(i)+ "  " + "   " +str(operazione))
```

```python
print_triangula_numbers(5)
```

# ESERCIZIO:  
Si scriva una funzione con un parametro intero n che calcola la somma dei numeri dispari nellintervallo (0, n).

```python
def SumOfOdds(n):
    somma = 0
    for i in range(n+1):
        if i % 2 != 0:
            somma = somma + i
            print(str(i) + "   " + str(somma))
    return(somma)   
        
```

```python
SumOfOdds(5)
```

# ESERCIZIO
Scrivere un programma che legga da tastiera un numero intero x e stampi il valore di x!. Per x! si intende x fattoriale

```python
def Fattoriale():
    x = int(input())
    operazione = 1
    for i in range(1, x+1):
        operazione = operazione * i
    return(operazione)
        
```
