# L'accumulazione

```python
def square(x):
    #inizializzo la variabile che poi mi dare lo square
    runningtotal = 0
    for i in range(1, x+1):
        runningtotal = runningtotal + x

    return runningtotal
```

```python
tosquare = 10
print(square(tosquare))
```
