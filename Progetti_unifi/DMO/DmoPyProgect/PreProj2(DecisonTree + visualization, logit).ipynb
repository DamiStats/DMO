```python
#Import dei paccheti per pre-processing
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport

# Import dei pacchetti per modelli di MACHINE LEARNING
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from IPython.display import Image  



# Import dei pacchetti per i modelli statistici
import statsmodels.api as sm
import statsmodels.formula.api as smf

#Import dataset:
melbourne_file_path = '/home/dami/Progetti_unifi/DMO/Dataset/melb_data.csv'
df = pd.read_csv(melbourne_file_path) 
del melbourne_file_path

#labello le colonnecolumns
```

```python
#labello le colonnecolumns
ciao = df.columns.values
type(df)
```

```python
#Scelgo covariate
y = df.Type
COVARIATE = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude','Price']
X = df[COVARIATE]
type(X)

df.dropna(axis=0)
type(y)
```

```python
# Binarizzo la y che ha 3 modalit√†
y.unique()
y.tolist
binary = []
for i in y:
    if i == "t" or i=="i":
        binary.append('t_u')
    else:
        binary.append('h')

binary
```

```python
y = binary
```

```python
#Imposto modelli
df_model_tree = DecisionTreeClassifier(criterion = 'gini', max_depth = 4).fit(X,y)
### Fitto modello
df_model_tree.fit(X, y)
```

```python
# Pacchetti per Per visualizzare l'albero
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

#
dot_data = StringIO()
export_graphviz(df_model_tree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
