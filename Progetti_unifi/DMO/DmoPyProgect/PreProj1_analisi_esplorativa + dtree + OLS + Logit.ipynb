```python
y = df.Type
COVARIATE = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude','Price']
X = df[COVARIATE]
```

# MODELLO (DECISION TREE ON TYPE):
- DEFINE
- FIT
- PREDICT
- EVALUATE

- DEFINE

```python
df_model_tree = DecisionTreeClassifier(criterion = 'gini', max_depth = 3).fit(X,y)

#random_state=1 indica il seed
```

- FIT

```python
df_model_tree.fit(X, y)
```

- PREDICT

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(df_model_tree.predict(X.head()))
```
